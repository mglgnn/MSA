```python
import numpy as np
from math import *

# p = 0.8 # probability of a match an any particular site
# k = 5   # required length of match
# L = 20   # length of sequence region
# Mario Stanke's version
def sensitivity1(p = 0.8, k = 5, L = 20):
    F = np.zeros([L+1, k+1])

    for i in range(k, L+1):
        for ell in range(k, -1, -1):
            if (ell == k):
                F[i, ell] = 1
            else:
                F[i, ell] = (1-p) * F[i - ell - 1, 0] + p * F[i, ell + 1]

    return F[L,0]

def sensitivity1Ex(p = 0.8, k = 5, L = 20):
    upper = L-k+1
    F = np.zeros([L+1, k+1, upper+1])

    for n in range(0, upper+1):
        for i in range(k, L+1):
            for ell in range(k, -1, -1):
                if (ell == k):
                    if n>0:
                        F[i, ell, n] = F[i-1, k-1, n-1]
                    else:
                        F[i, ell, n] = 1
                else:
                    F[i, ell, n] = (1-p) * F[i - ell - 1, 0, n] + p * F[i, ell + 1, n]

    return F[L,0]

def sensitivity1exNooverlap(p = 0.8, k = 5, L = 20):
    upper = 1 + floor((L+1)/(k+1))
    F = np.zeros([L+1, k+1, upper+1])

    for n in range(0, upper+1):
        for i in range(k, L+1):
            for ell in range(k, -1, -1):
                if (ell == k):
                    if n>0:
                        F[i, ell, n] = F[i-k, 0, n-1]
                    else:
                        F[i, ell, n] = 1
                else:
                    F[i, ell, n] = (1-p) * F[i - ell - 1, 0, n] + p * F[i, ell + 1, n]

    return F[L,0]


# main
sn1 = sensitivity1()
print("sensitivity :")
print("P(num matches >0) = ", sn1)
print()

sn1 = sensitivity1Ex()
print("extended sensitivity (overlap):")
for i in range(0,len(sn1)):
    print("P(num matches >= ", i, ") = ", sn1[i])
print()
sn1 = sensitivity1exNooverlap()
print("extended sensitivity (no overlap):")
for i in range(0,len(sn1)):
    print("P(num matches >= ", i, ") = ", sn1[i])
```
